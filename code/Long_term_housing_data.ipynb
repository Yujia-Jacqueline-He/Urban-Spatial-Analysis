```python
import pandas as pd
```

```python
Value = pd.read_csv("Parcels_2016_Data_Full.csv")
for idx in range(17):    

    Data = pd.read_csv("GeocodeResults ("+ str(idx+1) +").csv",header = None)

    if (idx+1)*9999 > Value.shape[0]:
        end = Value.shape[0]
    else:
        end = (idx+1)*9999

    dl = pd.DataFrame(Value["AV_BLDG"][idx*9999:end].reset_index(drop = True))
    dr = Data[[0,5,10]]

    value = pd.merge(dl, dr, left_index=True, right_on=0).drop(0, axis = 1)
    value = value[~value[10].isna()]
    value["CT"] = value[10].copy().astype(int).astype(str)

    value.drop(10,axis=1)[value.AV_BLDG != 0].to_csv("CT_Value_"+str(idx+1)+".csv")   
```
